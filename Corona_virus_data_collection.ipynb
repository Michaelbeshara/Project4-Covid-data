```python
import requests
import csv
```

```python
response = requests.get("https://coronavirus.m.pipedream.net/")
response
```

```python
corona = response.json()
corona
```

```python
corona['rawData']
```

```python
with open("COVID_Data.csv", 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['country_name', 'province_state', 'Confirmed', 'Deaths'])
    writer.writeheader()

    for c in corona['rawData']:
        country = c['Country_Region']
        province = c['Province_State']
        confirmed = c['Confirmed']
        deaths = c['Deaths']

        writer.writerow(
            {
                'country_name' : country,
                'province_state' : province,
                'Confirmed' : confirmed,
                'Deaths' : deaths
            }
        )
```
